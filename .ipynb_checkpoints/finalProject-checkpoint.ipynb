```python
import pandas as pd
import numpy as np

normalized_data = pd.read_csv("data.csv")
federal_awards = pd.read_csv("FederalCoCData.csv", low_memory=False)

# Shrinking the dataframe to avoid empty rows
federal_awards = federal_awards[:31992]
federal_awards['Year'] = federal_awards['Year'].astype('Int64') 

#droping unnecessary columns
federal_awards = federal_awards.drop(columns=federal_awards.columns[[1, 2, 5]])

federal_awards[federal_awards.columns[1]] = federal_awards[federal_awards.columns[1]].astype(str)
federal_awards = federal_awards.rename(columns={'Unnamed: 3': 'CoC Number'})
federal_awards[['Org Name', 'Program Type']] = federal_awards[['Org Name', 'Program Type']].astype(str)
federal_awards['Award Amount'] = federal_awards['Award Amount'].astype('Int64')
federal_awards_Final = federal_awards.groupby(['CoC Number', 'Year'])['Award Amount'].sum().reset_index()

# federal_awards[federal_awards.columns[2]] = federal_awards[federal_awards.columns[2]].astype(str)
# federal_awards[federal_awards.columns[3]] = federal_awards[federal_awards.columns[3]].astype(str)

# federal_awards[federal_awards.columns[4]] = federal_awards[federal_awards.columns[4]].astype(int)

# Pivot federal_awards_Final so each year becomes a column
federal_awards_Final_pivot = federal_awards_Final.pivot(index='CoC Number', columns='Year', values='Award Amount')
federal_awards_Final_pivot.columns = [f'Award Amount - {int(year)}' for year in federal_awards_Final_pivot.columns]
federal_awards_Final_pivot = federal_awards_Final_pivot.reset_index()

# Join onto df on CoC Number
df_final = normalized_data.merge(federal_awards_Final_pivot, left_on='CoC Number', right_on='CoC Number', how='left')

# print(df['CoC Number'].head())
# print(federal_awards_Final['CoC Number'].head())
df_final = df_final.dropna()

df_final = df_final.sort_values('CoC Number')

pd.set_option('display.max_columns', None)
df_final
```

```python
years = range(2010, 2025)

for year in years:
    award_col = f'Award Amount - {year}'
    homeless_col = f'Overall Homeless {year}'
    new_col = f'Award Per Unhoused Person - {year}'
    
    if award_col in df_final.columns and homeless_col in df_final.columns:
        df_final[new_col] = df_final[award_col] / df_final[homeless_col]
        # print(df_final[new_col])
    # else:
        # print(award_col, homeless_col, "DNE")

df_final.iloc[:, -20:]
```

```python
for col in df_final.columns:
    print(col)
```

```python
years = range(2010, 2025)

# Non-relative score: weighted sum of normalized chronic, overall, and unsheltered homeless
for year in years:
    chronic_col = f'Estimated Overall Chronically Homeless {year} Normalized'
    overall_col = f'Estimated Overall Homeless {year} Normalized'
    unsheltered_col = f'Estimated Unsheltered Homeless {year} Normalized'
    
    if all(c in df_final.columns for c in [chronic_col, overall_col, unsheltered_col]):
        df_final[f'Non-Relative Score {year}'] = (
            0.6 * df_final[chronic_col] +
            0.3 * df_final[overall_col] +
            0.1 * df_final[unsheltered_col]
        )

# Relative score: (current / prior) * current non-relative score
# 2010 base case: relative = non-relative
df_final['Relative Score 2010'] = df_final['Non-Relative Score 2010']

for year in range(2011, 2025):
    current_col = f'Non-Relative Score {year}'
    prior_col = f'Non-Relative Score {year - 1}'
    
    if all(c in df_final.columns for c in [current_col, prior_col]):
        df_final[f'Relative Score {year}'] = (
            df_final[current_col] / df_final[prior_col]
        ) * df_final[current_col]

df_final.iloc[:, -50:]
```
